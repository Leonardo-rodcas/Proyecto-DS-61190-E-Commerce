## Procesos y Subprocesos:

* Facturación de Servicios Logísticos

## **Proceso de Preparado de un pedido**
Comienza cuando el mismo, llega al Sistema ERP de gestión de los mismos y que pertenece a **"Epsilon S.A."** En este proceso, se puede diferenciar las siguientes etapas:
*   *Picking de materiales solicitados*: en este proceso, las personas encargadas de armar los pedidos, buscan en el depósito los materiales necesarios para esta tarea. Esto implica, movimientos internos entre los materiales que están almacenados en los racks, hacia los distintos puestos de preparado
*   *Contabilización de materiales*: este proceso, realiza los movimientos y ajuste de stock necesario en el Sistema ERP de **"Epsilon S.A."**
*   *Asignación de un transportista para el pedido*: en este se determina según distintas reglas de negocio, quién será el proveedor encargado de realizar la distribución del pedido
*   *Imposición del pedido en el proveedor*: en este proceso, el proveedor nos envía por Web Service un ID de seguimiento para cada pedido. Este ID, es informado a cada cliente para que puedan realizar el respectivo seguimiento de su compra.

## **Proceso de Distribución**
Comienza apenas un distribuidor retira de los Centros de Preparados, los pedidos que le han sido asignados. En este proceso, cada pedido puede tener los siguientes estados:
*   *En Curso*: el pedido se encuentra en poder del distribuidor
*   *Entregado*: el pedido ha sido entregado al cliente
*   *Devuelto*: el pedido ha sido devuelto al centro que preparó el pedido
*   *Siniestrado*: el pedido ha sufrido un robo o hurto
*   *Destruido*: los pedidos en los que solamente se solicitan "Sims", el proveedor si no los puede entregar al cliente, se encarga de la destrucción de este tipo de material

**"Epsilon S.A."** nos informa varias problemáticas a abordar, y nos proponen los siguientes proyectos (a desarrollar por etapas), porque se quieren apalancar de la Ciencia de Datos para mejorar sus procesos:

1. Predicción de Ingreso de Stock
2. Predicción de la demanda de Pedidos (próxima semana, próximo mes, próximo año)
3. Mejora de la Efectividad de la Primer Visita del Proveedor Logístico

Según el relevamiento realiza en cojunto, se ha decidido avanzar con el proyecto de **"Mejora de la Efectividad de la Primer Visita del Proveedor Logístico"**, por las siguientes razones:
* Mejorar la imagen ante nuestros CLIENTES
* Evitar reclamos teléfónicos/por mail al área comercial, por parte de los clientes, por demoras en el arribo de sus pedidos
* Reducir costos de distribución, ya que mejorando la tasa de entrega en la primer visita, vamos a:
  * Evitar segundas o terceras visitas, lo cual trae un costo extra aparejado
  * Evitar Devoluciones innecesarias de pedidos

#**Implementación del Modelo de Regresión Logístico**:
Se propone implementar un modelo de regresión, que nos permita decidir si una determinada fecha, es crítica. Para ello, se ha definido crítico a aquel día en el cual la cantidad de pedidos es superior a 4000

```python
# Hacemos una copia del modelo original
df_pedidos_mod = df_pedidos
```

```python
# Agrupamos por 'Centro', 'Canal', 'Subcanal' y 'Distribuidor' y aplicamos múltiples agregaciones
df_pedidos_mod = df_pedidos_mod.groupby(['Id_Tipo_Centro','Id_Canal','Id_Subcanal','Id_Distribuidor']).agg({
    'Cantidad_de_Pedidos': 'sum',
    'Cant_Celulares': 'sum',
    'Cant_Tarjetas_SIMs': 'sum',
    'Cant_Accesorios': 'sum',
    'Cant_Accesorios_IOT': 'sum'
    # Ejemplo de contar el número de fechas por grupo
}).reset_index()
print(df_pedidos_mod)
```

```python
# Creamos el campo "Target"
conditions = [
    (df_pedidos_mod['Cantidad_de_Pedidos'] <= 13),
    (df_pedidos_mod['Cantidad_de_Pedidos'] > 46) & (df_pedidos_mod['Cantidad_de_Pedidos'] < 106),
    (df_pedidos_mod['Cantidad_de_Pedidos'] >= 106)
    ]
values = [0,1, 2]
# Crear la columna 'Target' usando np.select
df_pedidos_mod['Target'] = np.select(conditions, values)
# Mostrar el dataframe con la nueva columna 'Target'
```

```python
# Definimos X
X = df_pedidos_mod[['Id_Tipo_Centro','Id_Canal','Id_Subcanal','Id_Distribuidor', 'Cant_Celulares','Cant_Tarjetas_SIMs','Cant_Accesorios','Cant_Accesorios_IOT']]
X
```

```python
# Definimos y
y = df_pedidos_mod['Target'].values
y
```

```python
#Crear la instancia del modelo
clf=LogisticRegression() #Ajustamos nuestro modelo a los datos de x e y
```

```python
# Entrenar el modelo con los datos y la etiqueta
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=45)
clf.fit(X_train, y_train)
```

```python
# Se realiza la predicción con el modelo entrenado
y_pred = clf.predict(X_test)
accuracy = clf.score(X_test,y_test)
print(f'La exactitud obtenida es de: { accuracy}')
```

```python
# Creamos la Matriz de Confusión
confusion_matrix(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
print(cm)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot()
plt.show()
```

#**Conclusiones del Modelo de Regresión Logístico**:
Se ha conseguido un buen accuracy (0.857). Según la matriz de confusión:
* Para el target = 0 (demanda baja) se ha obtenido un 100% de predicción
* Para el target = 1 (demanda media) solo hubo un caso y no fué bien clasificado
* Para el target = 2 (demanda alta) se ha obtenido una gran predicción (13 casos de 16)

Se concluye que es un buen modelo para deducir este target
